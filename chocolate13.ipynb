```python


print('⚠️ Calculate Outliers')
col = df["Boxes Shipped"]
```

```python
Q1 = col.quantile(0.25)
print("Q1 =", Q1)
```

```python
Q3 = col.quantile(0.75)
print("Q3 =", Q3)
```

```python
IQR = Q3 - Q1
print("IQR =", IQR)
print('⚠️ Outlier Limits') 
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print("Lower Bound =", lower_bound)
print("Upper Bound =", upper_bound)
```

```python
df['Amount'] = df['Amount'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
```

# 🍫 Chocolate KPIs

```python
# total revenue
total_sales = df['Amount'].sum()
total_boxes = df['Boxes Shipped'].sum()
print(f"Total Revenue: ${total_sales:,.2f} | Total Boxes: {total_boxes:,.2f}")
```

```python


df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.strftime('%B')
```

```python
# total revenue by yars and month
fig = px.sunburst(
    df, 
    path=['Year', 'Month'], 
    values='Amount',
    color='Amount',
    color_continuous_scale='Brwnyl',
    title='Sales Distribution by Year and Month (Sunburst View)'
)
fig.update_layout(
    margin=dict(t=50, l=0, r=0, b=0),
    width=600,
    height=600
)
fig.show()
```

```python
sales_by_country = df.groupby('Country')['Amount'].sum().sort_values(ascending=False)
print(f"📶 Sales by Country:, {sales_by_country}")
```

```python
plt.figure(figsize=(12, 6))
plot = sns.barplot(data=sales_by_country.reset_index(), x='Country', y='Amount', palette='YlOrBr_r')
plt.title('Total Sales Revenue by Country', fontsize=16, fontweight='bold')
plt.xlabel('Country', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)
```

```python
sales_by_person = df.groupby('Sales Person')['Amount'].sum().sort_values(ascending=False)
print(f"🔗 Top  Sales Persons:, {sales_by_person}")
```

```python
sales_by_product = df.groupby('Product')['Amount'].sum().sort_values(ascending=False)
print(f"🔗Top Products by Revenue:, {sales_by_product}")
```

```python
# Average Revenue per Box
avg_rev_per_box = total_sales / total_boxes
print(f"Average Revenue per Box: ${avg_rev_per_box:.2f}")
```

```python
country_avg = df.groupby('Country').apply(
    lambda x: x['Amount'].sum() / x['Boxes Shipped'].sum()
).sort_values(ascending=False).reset_index()

country_avg.columns = ['Country', 'Avg_Revenue_Per_Box']
```

```python
plt.figure(figsize=(12, 6))
plot = sns.barplot(data=country_avg, x='Avg_Revenue_Per_Box', y='Country', palette='copper')
plt.title('Average Revenue per Box by Country', fontsize=15, fontweight='bold')
plt.xlabel('Average Revenue ($)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
print(country_avg)
```
